Generando los valores extremos.

```python
np.random.seed(0)
data=np.random.normal(loc=0, scale=1, size=(10000,10))
for i in range (data.shape[1]):
    #Con esto agregaremos valores extremos positivos
    data[np.random.choice(range(tot_reg), size=10, replace=False), i] += np.random.uniform(10,20, size=10)

    #Con esto, agregamos valores extremos negativos.
    data[np.random.choice(range(tot_reg), size=10, replace=False), i] -= np.random.uniform(10,20, size=10)

df_bueno=pd.DataFrame(data, columns=[f"variable continua: {i+1}" for i in range (10)])
data
```

IQR

```python
q1=df_bueno.quantile(0.25)
q3=df_bueno.quantile(0.75)
```

```python
IQR=q3-q1
```

```python
mask=-(((df_bueno <(q1-1.5*IQR)) | (df_bueno > q3 +1.5*IQR))).any(axis=1)
```

```python
from scipy.stats import zscore
```

```python
z_scores= df_bueno.apply(zscore)
```

```python
mask_z = (z_scores > -3) & (z_scores < 3)
mask_z
```
